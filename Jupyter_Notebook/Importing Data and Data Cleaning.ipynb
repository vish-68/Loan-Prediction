### Data Cleaning

```python
#using isnull().sum() we can get how many null value is present in each variable:
df1.isnull().sum()
```

```python
#drop that variable whose missing value is 50% according to data entries:
round(((855969/100)*50),1)
```

```python
#using for loop for droping variable so we don't have to type manually for each variable:
for x in df1:
    if df1[x].isnull().sum()>427985:
        df1=df1.drop(x,axis=1)
df1.isnull().sum()
```

```python
#we will drop variables with help of correlation plot:
corr_df=df1.iloc[:,:].corr(method="pearson")

#change the size of heatmap:
plt.figure(figsize = (30,10))

sns.heatmap(corr_df,vmax=1.0,vmin=-1.0,annot=True,linewidths=.5)
```

```python
#we will drop those variable whose is multi-collinear, having single value :
df1=df1.drop(['id', 'member_id','funded_amnt', 'funded_amnt_inv','emp_title','pymnt_plan','title',
             'earliest_cr_line','total_pymnt_inv', 'recoveries', 'collection_recovery_fee',
             'last_pymnt_d','next_pymnt_d', 'last_credit_pull_d','policy_code',
             'total_rev_hi_lim','zip_code'],axis=1)
```

```python
#checking wheatherall multi-collinear variables are drop:
corr_df=df1.iloc[:,:].corr(method="pearson")

#change the size of heatmap:
plt.figure(figsize = (30,10))

sns.heatmap(corr_df,vmax=1.0,vmin=-1.0,annot=True,linewidths=.5)
```

```python
#printing column name and shape:
print(df1.columns)
print(df1.shape)
```

```python
#checking missing value again because we drop some variable:
df1.isnull().sum()
```

```python
#we will treat missing values according to their datatypes(object,boolean with mode,numerical with mean):
col_name=df1[['emp_length','revol_util','tot_cur_bal','collections_12_mths_ex_med','tot_coll_amt']]
for x in col_name:
    if col_name[x].dtypes=='object' or col_name[x].dtypes=="bool":
        df1[x].fillna(df1[x].mode()[0],inplace=True)
    elif col_name[x].dtypes=='int64' or col_name[x].dtypes=='float64':
        df1[x].fillna(df1[x].mean(),inplace=True)

df1.isnull().sum()
```

```python
#storing categorical value in 'df1_categorical' and numerical in 'df1_numerical'
df1_categorical=[]
df1_numerical=[]

for x in df1:
    if df1[x].dtype=='object':
        df1_categorical.append(x)
    elif df1[x].dtype=='int64' or df1[x].dtype=='float64':
        df1_numerical.append(x)
```

```python
#checking unique value in 'home_ownership' column:
df1['home_ownership'].unique()
```

```python
#replacing 'NONE' and 'ANY' with 'OTHER':
df1['home_ownership']=df1['home_ownership'].replace({'NONE':'OTHER','ANY':'OTHER'})
df1['home_ownership'].unique()
```

```python
#checking unique value in 'verification_status' column:
df1['verification_status'].unique()
```

```python
#replacing 'Source Verified' to 'Income Source Verified'
df1['verification_status'].replace({'Source Verified':'Income Source Verified'},inplace=True)
df1['verification_status'].unique()
```
