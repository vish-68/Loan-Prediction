### Label Encoding

```python
#Converting 'issue_d' from string to datetime object:
df1['issue_d'] = pd.to_datetime(df1['issue_d'])
df1.dtypes
```

```python
#storing categorical value in 'df1_categorical' and numerical in 'df1_numerical'
df1_categorical=[]
df1_numerical=[]

for x in df1:
    if df1[x].dtype=='object':
        df1_categorical.append(x)
    elif df1[x].dtype=='int64' or df1[x].dtype=='float64':
        df1_numerical.append(x)
```

```python
df1_categorical
```

```python
# For preprocessing the data 
from sklearn import preprocessing 
le=preprocessing.LabelEncoder()

for i in df1_categorical:
    df1[i]=le.fit_transform(df1[i])
    
    le_name_mapping=dict(zip(le.classes_, le.transform(le.classes_)))
    print('Feature',i)
    print('le_name_mapping',le_name_mapping)
    print()
```
