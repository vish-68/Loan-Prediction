```python
#importing libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#to avoid warnings:
import warnings
warnings.filterwarnings('ignore')

#to display all rows and columns:
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
#import dataset:
df=pd.read_csv(r'C:\Users\VISHAL\New folder\User Interface\data\exported_data\CleanData.csv',header=0)
df.head(10)
```

```python
#copy dataset in another file:
df1=df.copy()
df1.head(10)
```

```python
#no. of columns and rows
df1.shape
```

```python
#storing categorical value in 'df1_categorical' and numerical in 'df1_numerical'
df1_categorical=[]
df1_numerical=[]

for x in df1:
    if df1[x].dtype=='object':
        df1_categorical.append(x)
    elif df1[x].dtype=='int64' or df1[x].dtype=='float64':
        df1_numerical.append(x)
```

```python
sns.countplot(x='term',data=df1,palette="gist_rainbow_r",hue='grade')
plt.title("Term Gradewise")
```

```python
sns.countplot(x='home_ownership',data=df1,palette="gist_rainbow_r",hue='grade')
plt.title("Home Ownership Gradewise")
#plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='addr_state',data=df[df.default_ind==1],palette="gist_rainbow_r",order=df1['addr_state'].value_counts().sort_values(ascending=False).index)
plt.title("Defaulter by State")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='addr_state',data=df[df.default_ind==0],palette="gist_rainbow_r",order=df['addr_state'].value_counts().sort_values(ascending=False).index)
plt.title("Non-Defaulter by State")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(y='purpose',data=df,palette="gist_rainbow_r",order=df['purpose'].value_counts().sort_values(ascending=False).index)
plt.title("Purpose of Loan")
#plt.yticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='verification_status',data=df1,palette='gist_rainbow_r',hue="home_ownership",
              order=df1['verification_status'].value_counts().sort_values(ascending=False).index)
```

```python
j=1
x=10
plt.figure(figsize=(20,20))
for i in df1_numerical:
        #print(i)
        plt.subplot(10,10,j)
        sns.boxplot(data = df1, y= i)
        j=j+2
```

```python
j=1
y=10
plt.figure(figsize=(20,20))
for i in df1_numerical:
        #print(i)
        plt.subplot(10,10,j)
        plt.hist(data = df1, x= i)
        plt.title(i)
        j=j+3
```
